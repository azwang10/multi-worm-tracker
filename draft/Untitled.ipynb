## A Minimal "MultiWormTracker"

I don't like MATLAB, so I wrote this.

Given an input `.avi` file, this notebook will output the position for each worm at each frame in a `.csv` file. Does not extract any other features or extrapolate between frames where a worm is lost. Edit the parameters in ALL CAPS to fit to your data.

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'Arial'

#REQUIRES THREE LESS COMMON PACKAGES that can be installed with 'conda install scikit-image joblib ffmpeg'
from skimage import io, filters, measure, util
from joblib import Parallel, delayed
```

### 1 Locate files [EDIT]

The notebook requires a scratch directory with a lot of storage that will be filled with the `.avi` file converted into `.tif` for each individual frame. In my hands, 35 GB was used for a 45 min recording

```python
#edit
INPUT_AVI = '/Volumes/UNTITLED/Examples/03022023_N2_speed_3.avi'
SCRATCH_PATH = '/Volumes/UNTITLED'
OUTPUT_PATH = '/Volumes/UNTITLED'
FRAME_LIMIT = None

if not os.path.isfile(INPUT_AVI):
    raise FileNotFoundError()
```

### 2 Initialize scratch directory

```python
name_from_path = lambda name: os.path.splitext(os.path.basename(name))[0]
fname = name_from_path(INPUT_AVI)

scratch_dir = os.path.join(SCRATCH_PATH, fname + '-scratch')
os.makedirs(scratch_dir, exist_ok=True)
```

### 3 Convert `.avi` to `.tif` with `ffmpeg` (<5 min for a 45 min recording on my laptop)

```python
%%time

tif_dir = os.path.join(scratch_dir, 'tif')
os.makedirs(tif_dir, exist_ok=True)

def convert():
    if FRAME_LIMIT == None:
        os.system('ffmpeg -hide_banner -i %s -pix_fmt gray %s/%%06d.tif' % (INPUT_AVI, tif_dir))
    else:
        os.system('ffmpeg -hide_banner -i %s -pix_fmt gray -frames:v %i %s/%%06d.tif' % (INPUT_AVI, FRAME_LIMIT, tif_dir))

#comment the next line if you don't want to rerun it
#convert()
```

### 4 Generate a mask that's the mean out of an equally sampled 100 frames in the recording

```python
%%time

tif_list = sorted(glob.glob(os.path.join(tif_dir, '*.tif')))

def get_mean():
    sum_arr = np.zeros(io.imread(tif_list[0]).shape, np.float64)
    indexes = np.linspace(0, len(tif_list) - 1, 100, dtype=int)
    for i in range(len(indexes)):
        sum_arr += io.imread(tif_list[indexes[i]])
    mean_arr = util.img_as_ubyte(sum_arr / len(indexes) / 255)
    io.imsave(os.path.join(scratch_dir, 'mean.tif'), mean_arr)

#comment the next line if you don't want to rerun it
#get_mean()

#for plotting
mean_arr = io.imread(os.path.join(scratch_dir, 'mean.tif'))
plt.imshow(mean_arr, cmap='Greys_r', vmin=0, vmax=255);
```

### 5 Try out worm detection parameters first [EDIT if the plots look bad]

This cell first detects blobs on a frame by subtracting each frame from the mean masking frame above and then thresholds that difference frame. A gaussian filter is also applied in between (which may actually be unnessary, but I feel makes the blob detection more robust; it definitely is necessary if a worm is somehow split into two halves by the thresholding, where using the filter rejoins them). Additionally, the threshold should be chosen to be as stringent as possible so that not many blobs are detected in the first place. Having too many blobs greatly slows the code.

It then chooses which blobs are worms by just picking ones that are roughly the size of a worm and also uses an eccentricity cutoff remove any features that may be from the edge of the plate.

```python
#edit if needed
GAUSSIAN_SIGMA = 2
THRESHOLD = 10
MIN_AREA = 10
MAX_AREA = 200
MAX_ECCENTRICITY = 0.99

def get_worms(tif):
    frame = (mean_arr - io.imread(tif)).astype(np.byte)
    filtered_frame = filters.gaussian(frame, sigma=GAUSSIAN_SIGMA, preserve_range=True)
    label_img = measure.label(filtered_frame > THRESHOLD)
    props = pd.DataFrame(measure.regionprops_table(label_img, properties=['area', 'centroid', 'eccentricity']))
    worm_indexes = np.all([props['area'] > MIN_AREA, props['area'] < MAX_AREA, props['eccentricity'] < MAX_ECCENTRICITY], axis=0)
    worm_df = props.iloc[worm_indexes]
    return frame, label_img, filtered_frame, worm_df

def check_params(rows=5):
    fig, axs = plt.subplots(rows, 5, figsize=(15, 3 * rows))
    if rows == 1:
        axs = np.expand_dims(axs, 0)
    indexes = np.linspace(0, len(tif_list) - 1, rows).astype(int)
    for i in range(rows):
        tif = tif_list[indexes[i]]
        frame, label_img, filtered_frame, worm_df = get_worms(tif)
        axs[i, 0].imshow(frame, vmin=0, vmax=np.percentile(frame[frame > 0], 99))
        axs[i, 0].set_ylabel('Frame %i' % (indexes[i] + 1))
        axs[i, 1].imshow(filtered_frame, vmin=0, vmax=np.percentile(frame[frame > 0], 99))
        axs[i, 2].imshow(label_img > 0)
        axs[i, 3].imshow(np.isin(label_img, worm_df.index + 1))
        axs[i, 4].hist(worm_df['area'], label='%i blobs to %i worms' % (np.max(label_img), len(worm_df)))
        axs[i, 4].set_xlim(MIN_AREA, MAX_AREA)
        axs[i, 4].legend()
        axs[0, 0].set_title('Mean subtracted frame'); axs[0, 1].set_title('Gaussian filtered frame')
        axs[0, 2].set_title('Thresholded frame'); axs[0, 3].set_title('Probable worms')
        axs[0, 4].set_title('Worm sizes')
    plt.tight_layout()

check_params()
```

### 6 Detect worms for each frame (<6 min for a 45 min recording on my laptop)

```python
%%time

worm_dir = os.path.join(scratch_dir, 'worms')
os.makedirs(worm_dir, exist_ok=True)

def detect(tif):
    num = name_from_path(tif)
    _, _, _, worm_df = get_worms(tif)
    out = pd.DataFrame()
    out['frame'] = [int(num)] * len(worm_df)
    out[['x', 'y']] = worm_df[['centroid-1', 'centroid-0']].round(2).to_numpy()
    out['id'] = range(len(out))
    out.to_csv(os.path.join(worm_dir, num + '.csv'), index=False)

#comment the next line if you don't want to rerun it
#Parallel(n_jobs=-1)(delayed(detect)(i) for i in tif_list)

#for plotting
worm_list = sorted(glob.glob(os.path.join(worm_dir, '*.csv')))
plt.plot([len(pd.read_csv(i)) for i in worm_list])
plt.ylabel('Worms per frame'); plt.xlabel('Frame #');
```

### 7 Link `.csv` (<1 min for a 45 min recording on my laptop)

```python
%%time

#edit if needed
SEPERATION = 15

untrimmed = os.path.join(scratch_dir, 'untrimmed_tracks.csv')

def link(index, track_list):
    #renumbers ids so that all ids are unique
    df1, df2 = pd.read_csv(track_list[index]), pd.read_csv(track_list[index+1])
    ids1, ids2 = sorted(set(df1['id'])), sorted(set(df2['id']))
    df1['id'] = df1['id'].replace(dict(zip(ids1, range(len(ids1)))))
    df2['id'] = df2['id'].replace(dict(zip(ids2, range(len(ids1), len(ids1) + len(ids2)))))

    #get last frame of first df and first frame of second df
    sub_df1 = df1[df1['frame'] == np.max(df1['frame'])]
    arr1 = np.array(sub_df1[['x', 'y']])
    sub_df2 = df2[df2['frame'] == np.min(df2['frame'])]
    arr2 = np.array(sub_df2[['x', 'y']])

    #get a dict of ids to be replaced, every worm from the first df should only link to one worm of the second df
    diff = arr1[:, np.newaxis, :] - arr2[np.newaxis, :, :]
    D = np.sqrt(np.sum(diff ** 2, axis=2))
    replace_dict = {}
    while True:
        if np.min(D) > SEPERATION:
            break
        index2 = np.argmin(np.min(D, axis=0))
        index1 = np.argmin(D[:, index2])
        replace_dict[sub_df2['id'].iloc[index2]] = sub_df1['id'].iloc[index1]
        D[index1], D[:, index2] = np.inf, np.inf
    
    #replaces ids and overwrites files
    df2['id'] = df2['id'].replace(replace_dict)
    df = pd.concat([df1, df2])
    df.to_csv(track_list[index], index=False)
    os.remove(track_list[index+1])

def parallel():
    track_dir = os.path.join(scratch_dir, 'tracks')
    os.makedirs(track_dir, exist_ok=True)
    os.system('cp -r %s/*.csv %s' % (worm_dir, track_dir))
    while True:
        track_list = sorted(glob.glob(os.path.join(track_dir, '*.csv')))
        if len(track_list) == 1:
            break
        Parallel(n_jobs=-1)(delayed(link)(i, track_list) for i in range(0, (len(track_list) // 2) * 2, 2))
    os.system('mv %s %s' % (track_list[0], untrimmed))
    os.rmdir(track_dir)

def plot_track_len(df):
    unique_values, counts = np.unique(df['id'], return_counts=True)
    plt.hist(counts, bins = np.arange(0, np.max(counts), 10)); plt.yscale('log')
    plt.xlabel('Length of track in frames'); plt.ylabel('Count')
    
#comment the next line if you don't want to rerun it
#parallel()

#for plotting
untrimmed_df = pd.read_csv(untrimmed)
plot_track_len(untrimmed_df)
```

### 8 Trim out short tracks

Removes tracks that are too short and move too little

```python
#edit if needed
MIN_FRAMES_SEEN = 100
MIN_AREA_TRAVELED = 10

output_csv = os.path.join(OUTPUT_PATH, fname + '.csv')

def trim(untrimmed_df):
    unique_values, counts = np.unique(untrimmed_df['id'], return_counts=True)
    trimmed_df = untrimmed_df[np.isin(untrimmed_df['id'], unique_values[counts > MIN_FRAMES_SEEN])]
    bbox_area = lambda xs, ys: (np.max(xs) - np.min(xs)) * (np.max(ys) - np.min(ys))
    ids_to_save = []
    for i in sorted(set(trimmed_df['id'])):
        x, y = trimmed_df[trimmed_df['id'] == i]['x'], trimmed_df[trimmed_df['id'] == i]['y']
        if bbox_area(x, y) > MIN_AREA_TRAVELED:
            ids_to_save.append(i)
    trimmed_df = trimmed_df[np.isin(trimmed_df['id'], ids_to_save)]
    ids = sorted(set(trimmed_df['id']))
    trimmed_df['id'] = trimmed_df['id'].replace(dict(zip(ids, range(len(ids)))))
    trimmed_df.to_csv(output_csv, index=False)
    return trimmed_df

#comment the next line if you don't want to rerun it
trimmed_df = trim(untrimmed_df)

#for plotting
plt.subplot(211)
unique_values, counts = np.unique(trimmed_df['frame'], return_counts=True)
plt.plot(unique_values, counts)
plt.ylabel('Worms per frame'); plt.xlabel('Frame #')
plt.subplot(212)
plot_track_len(trimmed_df)
plt.tight_layout()
```
